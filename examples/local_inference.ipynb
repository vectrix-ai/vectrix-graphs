# Local Inference Example

This example demonstrates local inference using publicly available large language models (LLMs) such as LLAMA 3.1 (405 billion parameters). Due to the significant computational requirements, we didn't actually run the model locally on a laptop. Instead, we utilized the hosting service provided by TogetherAI for inference.

It's worth noting that compared to more specialized models like Anthropic's Claude, LLAMA 3.1 has some limitations in generating well-formatted Markdown responses. To address this, we've opted to output plain text responses in this notebook. 

Potential improvements for future iterations could include:
1. Implementing a post-processing module to convert LLAMA 3.1 outputs into valid Markdown.
2. Fine-tuning the model to directly generate properly formatted Markdown responses.

These enhancements would improve the visual presentation of the model's outputs within Jupyter notebooks.

```python
# Load packages from the src directory
import sys
from IPython.display import Markdown, display, Image
sys.path.append('../src')

from langchain_core.messages import HumanMessage
from vectrix_graphs import local_slm_demo

# Display the graph
display(Image(local_slm_demo.get_graph().draw_mermaid_png()))
```

```python
#Ask the question
input = [HumanMessage(content="Hello, nice to meet you!")]


# Run the graph
response = await local_slm_demo.ainvoke({"messages": input})
print(response['messages'][-1].content)
```
